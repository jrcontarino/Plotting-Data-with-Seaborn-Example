# Plotting with Seaborn using World Cup Data

## Import Packages and Data

```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv("WorldCupMatches.csv")
#print(df.head())
print(df.info())
```

There are 20 columns and 852 rows in this dataframe.  This dataset is from Kaggle.

## Plot Total Goals

```python
## Add Total Goals column to dataframe
df["Total Goals"] = df["Home Team Goals"] + df["Away Team Goals"]
#print(df.head())

sns.set_style("whitegrid")
sns.set_context("poster", font_scale = .6)
f, ax = plt.subplots(figsize = (12, 7))
ax = sns.barplot(data = df, x = "Year", y = "Total Goals")
ax.set_title("FIFA Average Total Goals by Year")
plt.show()
```

The highest amount of average goals was scored in 1954 and the least in 1990.
